```python
gates = [
    frozenset({"orange", "green"}),
    frozenset({"blue", "yellow"}),
    frozenset({"orange"}),
    frozenset({"blue"})
]
types = np.array([2, 1, 0, 2, 2, 2, 2, 3, 3, 2, 3, 3, 1, 2, 2, 1, 0, 0])
```

```python
minmax = []
for i, type1 in enumerate(types):
    equals = []
    not_equals = []
    for j, type2 in enumerate(types):
        if i != j:
            if type1 == type2:
                equals.append(cors[i, j])
            else:
                not_equals.append(cors[i, j])
    minmax.append(max(not_equals) - max(equals))
```

```python
import sklearn.neighbors
from sklearn.model_selection import KFold
import sklearn.metrics
```

```python
data = allbincounts
labels = np.array(types)
```

```python
import sklearn.linear_model
```

```python
knn_model = sklearn.neighbors.KNeighborsClassifier(algorithm="brute", metric="correlation", n_neighbors=1)
kf = KFold(n_splits=3, shuffle=True)

scores = []
for train, test in kf.split(data):
    knn_model.fit(data[train, :], labels[train])
    predicted = knn_model.predict(data[test, :])
    scores.append(sklearn.metrics.accuracy_score(labels[test], predicted))
print(scores)
print(np.mean(scores))
```

```python
def server_send(packet, adress="send"):
    requests.post(r"http://localhost:5000/" + adress,json=packet, timeout=5)

bins = [
    "orange",
    "blue",
    "yellow",
    "green"
]

initialData = {}
initialData["signal"] = "initialize"
initialData["bins"] = [
    {"id":i, "color":matplotlib.colors.cnames[name]} for i, name in enumerate(bins)
]
initialData["bincounts"] = [
    [] for i in gates
]

initialData["directions"] = [list(i) for i in gates]
initialData["frameids"] = [-1]

server_send(initialData, "initialize")
```

```python
allcounts = np.vstack([np.bincount(counts, minlength=2**(3*4)) for counts in master.processor.log.counts])

fig = figure(figsize=(20, 10))

colorbin_models = []
for i, color in enumerate(bins):
    typesoi = [i for i, type_color in enumerate(gates) if color in type_color]
    ballsoi = [t in typesoi for t in types]
    frameidsoi = [i for ids in pd.DataFrame(master.decisionmaker.balls).loc[ballsoi]["ids"] for i in ids]
    
    model = sklearn.linear_model.LogisticRegression()
    model.fit(allcounts, [frameid in frameidsoi for frameid in range(allcounts.shape[0])])
    
    plot(model.predict_proba(allcounts)[:, 1]*(i+1), color=color)
    
    colorbin_models.append(model)
```
